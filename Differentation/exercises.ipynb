1. Create 100 random numbers x from a uniform distribution using the random() np function. Plot these random numbers on the xy plane, and in a histogram. Find the average and standard deviation of these numbers.

```python
rand_x = np.random.rand(100)

plt.plot(rand_x)
plt.show()
plt.hist(rand_x, bins=20)
plt.show()

print(f"Average: {np.mean(rand_x)}, Std: {np.std(rand_x)}")
```

3. Define the function $f(x,m,k) = e^{-mx} cos(2\pi kx)$ and compute f(x) for three different combination of alues of m and k.
Plot the three functions on the same plot, using different colors and line types
Annotate the graph with the values of m and k for each function.

```python
from math import e, cos, pi
from numpy.typing import NDArray
def f(x:NDArray, m:int, k:int) -> NDArray:
    # return e**(-m*k) * cos(2*pi*k*x)
    return np.exp(-m*k) * np.cos(2*pi*k*x)

x = np.linspace(0, 1, 100)
y1 = f(x, 2, 3)
y2 = f(x, 1, 1)
y3 = f(x, 3, 2)

plt.plot(x, y1, label="m=2, k=3")
plt.plot(x, y2, label="m=1, k=1")
plt.plot(x, y3, label="m=3, k=2")
plt.legend()
plt.show()

```

4. Write code to solve a cubic equation $x^3 + x^2 +1 = 0$ using solve() from sympy. Determine whether the roots are real or complex.

```python
from sympy import symbols, Eq, solve

# Define the variable and the equation
x = symbols('x')
equation = Eq(x**3 + x**2 + 1, 0)

# Solve the equation
roots = solve(equation, x)

# Determine whether the roots are real or complex
for root in roots:
    if root.is_real:
        print(f"Root {root} is real.")
    else:
        print(f"Root {root} is complex.")
```
