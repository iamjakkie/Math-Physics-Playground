Evaluate symbolic derivative of the function $$f = asin(bt)$$ with respect to time t, and plot the function f(t) and its derivative f'(t) for a given range t=0 to t=6. Use the numerical values a=1, b=2.

```python
import mplcyberpunk

import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, sin, diff
from sympy.utilities.lambdify import lambdify

# set matplotlib theme to dark
plt.style.use('cyberpunk')
```

```python
a, b, t = symbols('a,b,t', real=True)

f = a * sin(b * t)
deriv = diff(f, t)
print(f"Symbolic derivative v=dv/dt: {deriv}")
tims = np.linspace(0,6,50)
y = lambdify(t, f.subs({a:1, b:2}))
plt.subplot(1,2,1)
plt.plot(tims, y(tims))
plt.xlabel('Time [s]')
plt.ylabel('y(t)')

u = deriv.subs({a:1, b:2})
v = lambdify(t, u)

plt.subplot(1,2,2)
plt.plot(tims, v(tims))
plt.xlabel('Time [s]')
plt.ylabel('Speed v(t)=dy/dt')
plt.tight_layout()
plt.show()
```
